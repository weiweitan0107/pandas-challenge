```python
# Dependencies and Setup
import pandas as pd
from pandas import Series,DataFrame

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
unique_value = purchase_data["SN"].nunique()
Total_data_df= pd.DataFrame({"Total":[unique_value]})
Total_data_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
data_df = pd.DataFrame({"Number of Unique Items":[purchase_data["Item ID"].count()],
                        "Average Price":[purchase_data["Price"].mean()],
                        "Number of Purchase":[purchase_data["Purchase ID"].count()],
                        "Total Revenue":[purchase_data["Price"].sum()]
                       })

data_df.round({"Average Price":2, "Total Revenue":2})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_purchase_data3 = purchase_data.groupby(['Gender']).nunique()

gender_purchase_data3['Total Count'] = gender_purchase_data3['SN']
gender_purchase_data3['Percentage of Players'] = gender_purchase_data3['Total Count']/unique_value

gender_purchase_data3['Percentage of Players'] = pd.Series(["{0: .2f}%".format(val * 100) for val in gender_purchase_data3['Percentage of Players']], index = gender_purchase_data3.index)
gender_purchase_data3 = gender_purchase_data3.sort_values(["Total Count", "Percentage of Players"], ascending=False)

gender_purchase_data3[['Total Count', 'Percentage of Players']]

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_purchase_data = purchase_data.groupby(['Gender']).count()
gender_purchase_data2 = purchase_data.groupby(['Gender']).sum()
gender_purchase_data['Purchase Count'] = gender_purchase_data['Purchase ID']
gender_purchase_data['Average Purchase Price'] = gender_purchase_data2['Price']/gender_purchase_data['Purchase ID']
gender_purchase_data['Total Purchase Value'] = gender_purchase_data2['Price']
gender_purchase_data['Avg Total Purchase per Person'] = gender_purchase_data2['Price']/gender_purchase_data3['Total Count']

gender_purchase_data['Average Purchase Price'] = gender_purchase_data['Average Purchase Price'].astype(float).round(2)
gender_purchase_data['Avg Total Purchase per Person'] = gender_purchase_data['Avg Total Purchase per Person'].astype(float).round(2)



gender_purchase_data['Average Purchase Price'] = gender_purchase_data['Average Purchase Price'].map("$ {:,.2f}".format)
gender_purchase_data['Total Purchase Value'] = gender_purchase_data['Total Purchase Value'].map("$ {:,.2f}".format)
gender_purchase_data['Avg Total Purchase per Person'] = gender_purchase_data['Avg Total Purchase per Person'].map("$ {:,.2f}".format)

gender_purchase_data[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', 'Avg Total Purchase per Person']] 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_purchase_data4 = purchase_data.groupby(['SN']).max()

Total_SN = age_purchase_data4["Age"].count()

print(age_purchase_data4["Age"].max())
print(age_purchase_data4["Age"].min())
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_purchase_data4['Age Group'] = pd.cut(age_purchase_data4["Age"], bins, labels=age_labels)

age_group = age_purchase_data4.groupby("Age Group").count()
age_group['Total Count'] = age_group['Age']
age_group['Percentage of Players'] = age_group['Total Count']/Total_SN

age_group['Percentage of Players'] = pd.Series(["{0: .2f}%".format(val * 100) for val in age_group['Percentage of Players']], index = age_group.index)

age_group[['Total Count', 'Percentage of Players']] 
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_senders = purchase_data.groupby(['SN']).sum()
top_senders['Total Purchase Value'] = top_senders['Price']
print(top_senders.head())

top_senders['Purchase Count'] = purchase_data.groupby(['SN']).count()['Purchase ID']
print(top_senders.head())

top_senders['Average Purchase Price'] = top_senders['Total Purchase Value'] / top_senders['Purchase Count']

top_senders['Average Purchase Price'] = top_senders['Average Purchase Price'].astype(float).round(2)
top_senders = top_senders.sort_values("Total Purchase Value", ascending=False)
print(top_senders.head())

top_senders['Average Purchase Price'] = top_senders['Average Purchase Price'].map("$ {:,.2f}".format)
top_senders['Total Purchase Value'] = top_senders['Total Purchase Value'].map("$ {:,.2f}".format)

top_senders[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']].head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_items = purchase_data.groupby(['Item ID', 'Item Name']).sum()
top_items
top_items['Total Purchase Value'] = top_items['Price']
print(top_items.head())

top_items['Purchase Count'] = purchase_data.groupby(['Item ID', 'Item Name']).count()['Purchase ID']

top_items['Item Price'] = top_items['Total Purchase Value'] / top_items['Purchase Count']
print(top_items.head())

top_items = top_items.sort_values("Purchase Count", ascending=False)

top_items['Item Price'] = top_items['Item Price'].map("$ {:,.2f}".format)
top_items['Total Purchase Value'] = top_items['Total Purchase Value'].map("$ {:,.2f}".format)
top_items[['Purchase Count', 'Item Price', 'Total Purchase Value']].head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
top_items = purchase_data.groupby(['Item ID', 'Item Name']).sum()
top_items
top_items['Total Purchase Value'] = top_items['Price']


top_items['Purchase Count'] = purchase_data.groupby(['Item ID', 'Item Name']).count()['Purchase ID']

top_items['Item Price'] = top_items['Total Purchase Value'] / top_items['Purchase Count']


top_items = top_items.sort_values("Total Purchase Value", ascending=False)

top_items['Item Price'] = top_items['Item Price'].map("$ {:,.2f}".format)
top_items['Total Purchase Value'] = top_items['Total Purchase Value'].map("$ {:,.2f}".format)
top_items[['Purchase Count', 'Item Price', 'Total Purchase Value']].head()
```
